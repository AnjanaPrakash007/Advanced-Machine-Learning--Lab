```python
import numpy as np
X=np.array([[0.5,1.0],[3.0,4.0],[5.0,6.0]])
y=np.array([0,0,1,1])
m=np.zeros_like(X[0])
c=0
lr=0.0001
epochs=50
def sigmoid(z):
    return 1/(1+np.exp(-z))
for epoch in range(1,epochs+1):
    for i in range(len(X)):
        gr_wrt_m=X[i]*(y[i]-sigmoid(np.dot(m.T,X[i]+c)))
        gr_wrt_c=y[i]-sigmoid(np.dot(m.T,X[i])+c)
        print(f"Epoch {epoch},data point{i}:Gradient w r t m-{gr_wrt_m},Gradientt w r t c-{gr_wrt_c}")
        m=m+lr*gr_wrt_m
        c=c+lr*gr_wrt_c
predictions=[]
for i in range(len(X)):
    z=np.dot(m,X[i])+c
    y_pred=sigmoid(z)
    if y_pred>=0.5:
        predictions.append(1)
    else:
        predictions.append(0)
print("Predictions:", predictions)
print("Optimum weights (m):", m)
print("Optimum bias (c):", c)        
    
```
